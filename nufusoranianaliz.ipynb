```python
import pandas as pd

df = pd.read_csv("oranson.csv")
```

```python
df.columns
```

```python
df['2017 Nüfusu']
```

```python
# Veri türlerini integer'a dönüştür
df['2017 Nüfusu'] = df['2017 Nüfusu'].astype(int)
df['ÜLKER'] = df['ÜLKER'].astype(int)

# Oranları hesapla ve yeni bir sütuna ekle
df['Oranlar'] = df['ÜLKER'] / df['2017 Nüfusu'] * 100

pd.set_option('display.max_rows', None)
print(df)
```

```python
# Oranları büyükten küçüğe sırala
sorted_df = df.sort_values(by='Oranlar', ascending=False)

print(sorted_df)
```

```python
# Oranları büyükten küçüğe sırala
sorted_df = df.sort_values(by='2017 Nüfusu', ascending=False)

print(sorted_df)
```

```python
# Oranları büyükten küçüğe sırala
sorted_df = df.sort_values(by='ÜLKER', ascending=False)

print(sorted_df)
```
